```python
# Importing pandas to read .csv file
import pandas as pd
data = pd.read_csv('lsd_math_score_data.csv')
```

```python
# Print DataFrama obtained
print(data)
```

```python
# Importing libraries to notebook
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# Storing data
time = data[['Time_Delay_in_Minutes']]
LSD = data[['LSD_ppm']]
score = data[['Avg_Math_Test_Score']]
```

```python
%matplotlib inline

# Creating and customizing data graph
plt.title('Tissue concentration of LSD over time', fontsize=17) # Title of graph
plt.xlabel('Time in Minutes', fontsize=14) # X axis label
plt.ylabel('Tissue LSD ppm', fontsize=14) # Y axis label
plt.text(x=0, y=-0.5, s='Wagner et al. (1968)', fontsize=12) # Additional text
plt.xticks(fontsize=14) # Change size of x axis ticks
plt.yticks(fontsize=14) # Change size of y axis ticks

plt.ylim(1,7) # Set limits of y axis
plt.xlim(0,500) # Set limits of x axis

plt.style.use('classic') # Use style

plt.plot(time, LSD, color='#e74c3c', linewidth=3) # Plot data in graph and customize appearance
plt.show() # Show the graph
```

```python
# Use scikitlearn to create Linear Regression
regr = LinearRegression()
regr.fit(LSD, score)

# Print values obtained
print('Theta1 : ', regr.coef_[0][0]) 
print('Intercept: ', regr.intercept_[0])
print('R-Square: ', regr.score(LSD, score))

# Prediction
predicted_score = regr.predict(LSD)
```

```python
%matplotlib inline

# Creating and customizing graph
plt.title('Arithmetic vs LSD-25', fontsize=17) # Title of graph
plt.xlabel('Tissue LSD ppm', fontsize=14) # X axis label
plt.ylabel('Performance Score', fontsize=14) # Y axis label
plt.ylim(25, 85) # Set limits of y axis
plt.xlim(1, 6.5) # Set limits of x axis

plt.style.use('fivethirtyeight') # Use style

plt.scatter(LSD, score, color='blue', s=100, alpha=0.7) # Axis and transparency
plt.plot(LSD, predicted_score, color='red', linewidth=3) # Plot of prediction line
plt.show() # Show graph
```
