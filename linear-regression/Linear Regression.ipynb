```python
# Importing necessary libraries to notebook
import pandas
from pandas import DataFrame
import matplotlib.pyplot as plot
from sklearn.linear_model import LinearRegression
```

```python
# Reading the data from CSV file
data = pandas.read_csv('cost_revenue.csv')
```

```python
# Statistics about the read data
data.describe()
```

```python
# Creating DataFrames for each of the axis
X = DataFrame(data, columns=['production_budget_usd']) 
y = DataFrame(data, columns=['worldwide_gross_usd'])
```

```python
# Creating and customizing data graph 
plot.figure(figsize=(10,6)) # Size of graph
plot.scatter(X, y, alpha=0.3) # Axis and transparency
plot.title('Film Cost vs Global Revenue') # Title
plot.xlabel('Production Budget ($)') # Label of x axis
plot.ylabel('Worldwide Gross ($)') # Label of y axis
plot.ylim(0, 3000000000) # Limits of y axis
plot.xlim(0, 450000000) # Limits of x axis
plot.show() # Show the graph
```

```python
# Using scikitlearn to create Linear Regression
regression = LinearRegression()
regression.fit(X, y)
```

```python
# Slope coefficient (theta_1)
regression.coef_ 
# For each $1.00 spent on producing we should receive $3.11 of revenue in return
```

```python
# Intercept (theta_0)
regression.intercept_ 
# A movie with a budget of $0.00 would lose $7.2M ? no
```

```python
# Goodness of Fit (r)
regression.score(X, y)
# Amount of variation in the films revenue that is explained by the films budget
# Simple model of a single feature can explain around 55% of the variation
```

```python
# Creating new graph to show Linear Regression
plot.figure(figsize=(10,6)) # Size of graph
plot.scatter(X, y, alpha=0.3) # Axis and transparency
plot.plot(X, regression.predict(X), color='red', linewidth=3) # Plot of prediction line
plot.title('Film Cost vs Global Revenue') # Title
plot.xlabel('Production Budget ($)') # Label of x axis
plot.ylabel('Worldwide Gross ($)') # Label of y axis
plot.ylim(0, 3000000000) # Limits of y axis
plot.xlim(0, 450000000) # Limits of x axis
plot.show() # Show the graph
```
